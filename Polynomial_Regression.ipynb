## Polynomial Regression

**Importing the libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**Importing the dataset**

```python
data = pd.read_csv('position_salaries.csv')
data.head()
```

```python
X=data.iloc[:,1:-1].values
y=data.iloc[:,-1].values
```

```python
X
```

**Training Linear Regression model on the whole dataset**

```python
from sklearn.linear_model import LinearRegression
```

```python
lin_regressor= LinearRegression()
```

```python
lin_regressor.fit(X,y)
```

**Training the Polynomial model on the whole dataset**

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(4)
X_poly = poly.fit_transform(X)
```

```python
X_poly
```

```python
poly_regressor= LinearRegression()
```

```python
poly_regressor.fit(X_poly,y)
```

**Visualising the Linear Regression results**

```python
plt.scatter(X,y,c='red')
plt.plot(X,lin_regressor.predict(X))
plt.title('Truth or Bluff(Linear Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary');
```

**Visualising the Polynomial Regression**

```python
plt.scatter(X,y,c='red')
plt.plot(X,poly_regressor.predict(X_poly))
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary');
```

**Visualising the Polynomial Regression (for higher resolution and smoother curve)**

```python
X_lin=np.linspace(X.min(),X.max(),100).reshape(-1,1)
X_poly1 = poly.fit_transform(X_lin)
plt.scatter(X,y,c='red')
plt.plot(X_lin,poly_regressor.predict(X_poly1))
plt.title('Truth or Bluff(Polynomial Regression)')
plt.xlabel('Position Level')
plt.ylabel('Salary');
```

**Predicting a new result with Linear Regression**

```python
X_test = 6.5
lin_regressor.predict([[X_test]])
```

**Predicting a new result with Polynomial Regression**

```python
x_transformed = poly.transform([[X_test]])
poly_regressor.predict(x_transformed)
```
